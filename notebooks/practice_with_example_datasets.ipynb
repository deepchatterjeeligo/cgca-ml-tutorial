# Get some practice

The `sklearn` library provides some small example datasets to play around with. The data might not be particularly interesting for astronomy, but let's you get some practice. Here is an example dataset where different types of wines are classified in 3 categories.

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import datasets
```

```python
wine = datasets.load_wine()
```

```python
df = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])
```

The dataset has many features, let's focus on only a few and make a pairplot to take a look at the distributions

```python
g = sns.pairplot(df,
                 vars=['alcohol', 'magnesium', 'hue'],
                 hue='target',
                 diag_kind='auto')
```
